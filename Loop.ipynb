# PROGRAM-4

```python
a=int(input("Enter number upto which uh want to check odd and even 0:"))
counte=0
counto=0
for i in range (1,a+1,1):
    if (i% 2==0):
        counte +=1
    else:
        counto +=1
print(f"Total even number in series {counte} ")  
print(f"Total even number in series {counto} ") 
```

# PROGRAM -7

```python
n= int (input("Enter number uh want to check prime or not:"))
for i in range (2,(n//2)+1,1):
    if (n % i ==0 ):
        print("NOT PRIME")
        break
else:
    print("PRIME")
# else block will run if for loop does not break       
        
```
